**Подмечаем пропуски во многих столбцах**.

Для некоторых пропущенных значений можно предположить логичную замену. Например, если человек не указал число балконов — скорее всего, их нет. **Заменяем на 0:**

```python
data['balcony'].isna().sum() # проверяем замену
```

* Добавим функцию для категоризации этажа квартиры (например - первый, последний, другой):

```python
data['floors_total'].describe()
```

```python
data['floor'].describe()
```

```python
def floor_category(i): # называем функцию соответствующую действию, даем переменную i
    floors_total = i['floors_total'] # передаем переменной значения из таблицы
    floor = i['floor']
    
    if floor == 1: # делаем категоризацию
        return 'первый'
    elif floors_total == floor:
        return 'последний'
    else:
        return 'другой'

data['floor_category'] = data.apply(floor_category, axis=1) # применяем метод apply ко всему датафрейму
# методом axis даем пройтись по каждой строке датафрейма 
data.head()
```

* `column` - колонка для которого будет приниматься и делаться гистограмма;
* `bins` - ячейки в гистограмме;
* `range1` - границы для столбца column;
* `figsize` - размер гистограммы;

```python
def my_func(column, bins, figsize, range1):
    
    ax = local_spb.plot(kind='hist', 
                        y=column, 
                        histtype='step', 
                        range=range1, 
                        bins=bins, 
                        linewidth=2,
                        figsize=figsize,
                        alpha=0.5, 
                        label='вся по городу' + ' ' + column, 
                        grid=True,
                        density = True,
                        legend=True)
    
    local_spb_centralflats.plot(kind='hist', 
                                y=column, 
                                histtype='step', 
                                range=range1, 
                                bins=bins, 
                                linewidth=2,
                                figsize=figsize,
                                alpha=0.5, 
                                label= 'центра' + ' ' + column,
                                ax = ax,        # оси, по которым строится график
                                grid=True,
                                density = True, # это указывает на результат вычислений функции, то есть плотность вероятности на одном графике
                                legend=True
                                )
    
```
