# Технологический процесс

Как золото получают из руды? Нужно внимательно изучить стадии процесса.
Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

**Опишем каждую стадию:**


1. `Флотация`

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

2. `Очистка`

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

**Наименование признаков**

Наименование признаков должно быть такое:

**`[этап].[тип_параметра].[название_параметра]`**

**Пример:** `rougher.input.feed_ag`

**Возможные значения для блока**`[этап]`:
* `rougher` — флотация;
* `primary_cleaner` — первичная очистка;
* `secondary_cleaner` — вторичная очистка;
* `final` — финальные характеристики

Возможные значения для блока `[тип_параметра]`:
* `input` — параметры сырья;
* `output` — параметры продукта;
* `state` — параметры, характеризующие текущее состояние этапа;
* `calculation` — расчётные характеристики

**Описание данных**

`Технологический процесс`
*  `Rougher feed` — исходное сырье
* `Rougher additions (или reagent additions)` — флотационные реагенты: Xanthate, Sulphate, Depressant
   * `Xanthate` **— ксантогенат (промотер, или активатор флотации);
   * `Sulphate` — сульфат (на данном производстве сульфид натрия);
   * `Depressant` — депрессант (силикат натрия).
   
   
* `Rougher process (англ. «грубый процесс»)` — флотация;
* `Rougher tails` — отвальные хвосты;
* `Float banks` — флотационная установка;
* `Cleaner process` — очистка;
* `Rougher Au` — черновой концентрат золота;
* `Final Au` — финальный концентрат золота

**Параметры этапов**

* `air amount` — объём воздуха;
* `fluid levels` — уровень жидкости;
* `feed size` — размер гранул сырья;
* `feed rate` — скорость подачи

**Нужно спрогнозировать сразу две величины:**

* эффективность обогащения чернового концентрата **`rougher.output.recovery`**;
* эффективность обогащения финального концентрата **`final.output.recovery`**

### Проверяем, что эффективность обогащения рассчитана правильно. Вычисляем её на обучающей выборке (train) для признака `rougher.output.recovery` и находим `MAE` между нашими расчётами и значением признака

**Расчёт эффективности находится так:**



где:
* `C` — доля золота в концентрате после флотации/очистки;
* `F` — доля золота в сырье/концентрате до флотации/очистки;
* `T` — доля золота в отвальных хвостах после флотации/очистки

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат

Прибегая к данным, которые находятся в технологическом процессе, необходимые признаки под расчет этой формулы это:
* `С` - rougher.output.concentrate_au;
* `F` - rougher.input.feed_au; 
* `T` - rougher.output.tail_au

**Сначала проверим нужные нам колонки на пропуски и подставим эти значения для проверки формулы**

```python
def isna_for_recovery(column, columns): # аргумент columns несет описательный характер
    print(f'Кол-во пропусков в {columns}: {column.isna().sum()}')

isna_for_recovery(train['rougher.output.concentrate_au'], 'C')
isna_for_recovery(train['rougher.input.feed_au'], 'F')
isna_for_recovery(train['rougher.output.tail_au'], 'T')
display(train['rougher.output.recovery'].isna().sum()) # что логично, если нет пропусков по данным, по которым рассчитывается
```

```python
C = train['rougher.output.concentrate_au']
F = train['rougher.input.feed_au']
T = train['rougher.output.tail_au']

for_checking = pd.DataFrame(train['rougher.output.recovery']) # создаем датафрейм для улучшения визуального сравнения 

recovery_control = (C * (F - T)) / (F * (C - T)) * 100 # подставляем значения в формулу для проверки 

for_checking['recovery_control'] = recovery_control # создаем новый столбец в датафрейме для демонстрации сравнения
for_checking.tail(15)
```

На первый взгляд эффективность обогащения рассчитана правильно. 

Теперь рассчитаем **MAE** между нашими расчётами `(recovery_control)` и значением признака `(rougher.output.recovery)`:

```python
print('MAE между нашими расчётами и значением признака:', mean_absolute_error(recovery_control, 
                                                                              train['rougher.output.recovery']))
```

#### Вывод

Значение **MAE** (среднее абсолютное отклонение) **очень мало**, что говорит нам о **корректно** рассчитанных данных

### Анализ признаков, недоступные в тестовой выборке

Создадим пустой список `not_in_test`, куда будем помещать столбцы, которых нет в тестовой выборке:

```python
not_in_test = []
count = 0

for i in train.columns:
    if i not in test.columns:
        count += 1
        not_in_test.append(i)

display(train[not_in_test].head())
print('Кол-во признаков, которых нет в тестовой выборке', count, sep=': ')
#('; '.join(not_in_test))
not_in_test
```

**Подмечаем**, что все эти столбцы (признаки) получаются путем технологического процесса, то есть напрямую связан с нашими целевыми признаками. Выходной продукт рассчитывает целевой признак **recovery** на разных этапах. Иными словами они как бы сами целевые признаки, нельзя их интерпретировать как признаки

* Они описывают результаты процессов, таких как:
  * финальные характеристики того или иного продукта `(final.output)`, 
  * параметры продукта после первичной чистки `(primary_cleaner.output)`, 
  * расчетные характеристики флотации `(rougher.calculation)`, 
  * параметры продукта после флотации `(rougher.output)`, 
  * параметры продукта после второй чистки `(secondary_cleaner.output)`
   
При построении/обучении модели для спрогнозирования данных нам в задаче величин, они определенно будут только мешать, поэтому они не понадобятся на обучающей выборке. 

**Устраним эти признаки на обучающей выборке, а также добавим целевые признаки в тестовую выборку**

В нашей переменной `not_in_test` содержатся признаки, которых нет в тестовой выборке. Они нам и не нужны в обучающей, но там содержатся целевые признаки. Удалим их:

```python
display(train.loc[:, ['date', 'rougher.output.recovery', 'final.output.recovery']]) # перед удалением
print(f'Кол-во признаков перед удалением: {train.shape[1]}')
```

### Сравнение распределения размеров гранул сырья на обучающей и тестовой выборках

Также создадим функцию, которая будет демонстрировать визуальное сравнение, а также опишет этапы для каждой из двух выборок и посчитает медиану

```python
def feed_size(column, stage): # аргументы - признак с какой-либо очисткой (column), описательный этап очистки (stage) 
    
    x = train.loc[:, [column]] #  x, y - датафреймы с определенным этапом очистки на двух выборках
    y = test.loc[:, [column]]

    plt.figure(figsize=[14, 9])
    plt.hist(x, histtype='step', bins=70, label='train sample')
    plt.hist(y, bins=70, alpha=0.8, label='test sample')
    plt.title('Сравнение распределений размеров гранул сырья на этапе' +' '+ stage, fontweight='bold')
    plt.legend(fontsize=15)
    plt.show();
    
    print(f'Описание размера гранул сырья на этапе {stage} train выборка: / {x.describe()}')
    print()
    print(f'Медиана этапа {stage} train : {x.median()}')
    print()
    print(f'Описание размера гранул сырья на этапе {stage} test выборка: / {y.describe()}')
    print()
    print(f'Медиана этапа {stage} test: {y.median()}')
    
feed_size('rougher.input.feed_size', 'флотации')
```

## Модель

Для начала напишем функцию для вычисления sMAPE

Вычисляется она так:



**Строим функцию**

```python
def smape(target, predict):
    return 1 / len(target) * sum( abs(target - predict) / ( (abs(target) + abs(predict)) * 0.5) ) * 100
```

И также введем функцию итогового sMAPE, которая рассчитывается так:



```python
def final_smape(rougher, final):
    return 0.25 * rougher + 0.75 * final
```

Теперь присвоим признаки и таргеты для выборок. Они будут обучены на одном датасете (имеется в виду для флотации и финальной чистки). Никакого вреда, ровно как и пользы не будет. Так что можно обучить их на одних признаках

```python
features_train = train.drop(['date', 'rougher.output.recovery', 'final.output.recovery'], axis=1)
target_rougher = train['rougher.output.recovery'] 
target_final = train['final.output.recovery']
```

```python
features_train
```

* Создаем функцию для рассчитывания параметров и лучшего значения для **sMAPE**. Считать будем при помощи **`GridSearchCV()`**. Эта техника обучает модель с заданными нами в перечни парметрами, из которых выберет лучшие для метрики **sMAPE**, используя кросс валидацию (как раз она нам и нужна). 

* С помощью параметра `make_scorer()` внесем расчетную метрику - нашу **функцию smape**, а также параметр `greater_is_better=False`, которая будет считать скорер, чем меньше, тем лучше. Так как на выходе получается отрицательное число, приводим его в абсолютное с помощью `abs()`, чтобы не посчитал иначе при использовании этого параметра. 

* Лучшее значение **smape** выводим с помощью атрибута `.best_score_`, а лучшие параметры с помощью `.best_params_`

На вход фунцкия принимает **6 аргументов**: `сама модель`, `перечень параметров`, `признаки`, `целевой признак`, `название этапа` и `название модели`

```python
def best_model(model, params, features, target, name, model_name):
    
    cvs = GridSearchCV(model, params, scoring=make_scorer(smape, greater_is_better=False), cv=5)
    
    cvs.fit(features, target)
    
    print(f'Лучшие параметры на {name} этапе модели {model_name}: {cvs.best_params_}')
    print()
    print(f'Лучший sMAPE на {name} этапе модели {model_name}: {round(abs(cvs.best_score_), 4)}')
    print()    
```
